1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 10 * x - 5

def derivative(f, x, h=1e-5):
    return (f(x + h) - f(x)) / h

x = np.linspace(0, 10, 100)
y = f(x)
df = derivative(f, x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label="f(x) = 10x - 5", color="blue")
plt.plot(x, df, label="f'(x) = 10", color="red", linestyle="--")
plt.title("Функція та її похідна")
plt.xlabel("Кількість годин (x)")
plt.ylabel("Оцінка")
plt.legend()
plt.grid()
plt.show()
```

```python
import sympy as sp

x = sp.Symbol('x')
f = 10 * x - 5
df = sp.diff(f, x)

x_vals = np.linspace(0, 10, 100)

f_lambda = sp.lambdify(x, f, 'numpy')  # Перетворення функції на числову
df_lambda = sp.lambdify(x, df, 'numpy')

y_vals = f_lambda(x_vals)  # Обчислення значень f(x) для всіх x_vals
df_vals = np.full_like(x_vals, df_lambda(0))  # Створення масиву постійного значення похідної

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label="f(x) = 10x - 5", color="blue")
plt.plot(x_vals, df_vals, label="f'(x) = 10", color="red", linestyle="--")
plt.title("Функція та її похідна")
plt.xlabel("Кількість годин (x)")
plt.ylabel("Оцінка")
plt.legend()
plt.grid()
plt.show()
```

Функція f(x) = 10x - 5 є лінійною, а її похідна f'(x) = 10 показує постійну швидкість зміни оцінки. Це означає, що кожна додаткова година навчання додає 10 балів до оцінки.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
import sympy as sp

x = sp.Symbol('x')
f = sp.sin(x)
df = sp.diff(f, x)

x_vals = np.linspace(0, 10, 100)

f_lambda = sp.lambdify(x, f, 'numpy')  # Перетворення функції на числову
df_lambda = sp.lambdify(x, df, 'numpy')

y_vals = f_lambda(x_vals)  # Обчислення значень f(x) для всіх x_vals
df_vals = df_lambda(x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label="f(x) = sin(x)", color="blue")
plt.plot(x_vals, df_vals, label="f'(x)", color="red", linestyle="--")
plt.title("Функція та її похідна")
plt.legend()
plt.grid()
plt.show()
```

```python
import sympy as sp

x = sp.Symbol('x')
f = sp.ln(x)
df = sp.diff(f, x)

x_vals = np.linspace(0.1, 10, 100)

f_lambda = sp.lambdify(x, f, 'numpy')  # Перетворення функції на числову
df_lambda = sp.lambdify(x, df, 'numpy')

y_vals = f_lambda(x_vals)  # Обчислення значень f(x) для всіх x_vals
df_vals = df_lambda(x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label="f(x) = ln(x)", color="blue")
plt.plot(x_vals, df_vals, label="f'(x) = 1/x", color="red", linestyle="--")
plt.title("Функція та її похідна")
plt.legend()
plt.grid()
plt.show()
```

```python
import sympy as sp

x = sp.Symbol('x')
f = x**2 - 5*x + 1
df = sp.diff(f, x)

x_vals = np.linspace(0, 100, 500)

f_lambda = sp.lambdify(x, f, 'numpy')  # Перетворення функції на числову
df_lambda = sp.lambdify(x, df, 'numpy')

y_vals = f_lambda(x_vals)  # Обчислення значень f(x) для всіх x_vals
df_vals = df_lambda(x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label="f(x) = x^2 - 5x + 1", color="blue")
plt.plot(x_vals, df_vals, label="f'(x)", color="red", linestyle="--")
plt.title("Функція витрат і її похідна")
plt.xlabel("Кількість вироблених одиниць")
plt.ylabel("Витрати")
plt.legend()
plt.grid()
plt.show()
```
